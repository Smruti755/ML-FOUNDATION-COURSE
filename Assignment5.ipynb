```python
x = np.arange(0, 10)
a1 = np.random.randint(1, 100, size=10)
b1 = np.random.randint(1, 100, size=10)
a2 = np.random.randint(1, 100, size=10)
b2 = np.random.randint(1, 100, size=10)

# Create a 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Scatter plot of Dataset A
axs[0, 0].scatter(x, a1, color='blue', label='Dataset A', marker="o", s=50)
axs[0, 0].set_title("Scatter Plot of Dataset A")
axs[0, 0].set_xlabel("X-axis")
axs[0, 0].set_ylabel("Y-axis")
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: Scatter plot of Dataset B
axs[0, 1].scatter(x, b1, color='red', marker="*", s=100, label='Dataset B')
axs[0, 1].set_title("Scatter Plot of Dataset B")
axs[0, 1].set_xlabel("X-axis")
axs[0, 1].set_ylabel("Y-axis")
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot 3: Scatter plot of Dataset C
axs[1, 0].scatter(x, a2, color='green', label='Dataset C', marker="^", s=50)
axs[1, 0].set_title("Scatter Plot of Dataset C")
axs[1, 0].set_xlabel("X-axis")
axs[1, 0].set_ylabel("Y-axis")
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot 4: Scatter plot of Dataset D
axs[1, 1].scatter(x, b2, color='orange', marker="v", s=100, label='Dataset D')
axs[1, 1].set_title("Scatter Plot of Dataset D")
axs[1, 1].set_xlabel("X-axis")
axs[1, 1].set_ylabel("Y-axis")
axs[1, 1].legend()
axs[1, 1].grid(True)

# Layout so plots do not overlap
fig.tight_layout()

plt.show()
```

### Exercise 29
**Task:** Draw a filled area under a line (fill_between).

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.figure(figsize=(6,5))
plt.plot(x, y, label='sin(x)')
plt.title('Sine Wave')
plt.xlabel('Angle (radians)')
plt.ylabel('sin(x)')
plt.grid(True)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], 
           ['$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$'],
           rotation=45)
plt.yticks([-1, 0, 1], ['$Min$', '$0$', '$Max$'])

# Fill area under the curve
plt.fill_between(x, 0, y, where=(y >= 0), color='lightblue', alpha=0.3, label='Positive area')
plt.fill_between(x, 0, y, where=(y <= 0), color='red', alpha=0.3, label='Negative area')

plt.legend()
plt.tight_layout()
plt.show()
```

### Exercise 31
**Task:** Plot a stacked bar chart.

```python
labels = ['A', 'B', 'C', 'D']
series1 = [10, 15, 7, 12]
series2 = [8, 9, 13, 10]
series3 = [12, 10, 8, 15]
fig, ax = plt.subplots()

# Plot the stacked bar chart
ax.bar(labels, series1, label='Series 1')
ax.bar(labels, series2, bottom=series1, label='Series 2')
ax.bar(labels, series3, bottom=np.add(series1, series2), label='Series 3')
ax.set_title('Stacked Bar Chart')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.legend()
plt.show()
```

### Exercise 32
**Task:** Plot grouped bar charts.

```python
labels = ['A', 'B', 'C', 'D']
series1 = [10, 15, 7, 12]
series2 = [8, 9, 13, 10]
series3 = [12, 10, 8, 15]
bar_width = 0.25
# Set positions of the bars on the x-axis
x = np.arange(len(labels))
pos1 = x - bar_width
pos2 = x
pos3 = x + bar_width
fig, ax = plt.subplots()

# Plot the grouped bar chart
ax.bar(pos1, series1, width=bar_width, label='Series 1')
ax.bar(pos2, series2, width=bar_width, label='Series 2')
ax.bar(pos3, series3, width=bar_width, label='Series 3')
ax.set_title('Grouped Bar Chart')
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

### Exercise 40
**Task:** Create a 3D plot using matplotlib.

```python
# Create a figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data for the plot
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

# Show the plot
plt.show()
```

### Exercise 42
**Task:** Use color maps in heatmaps.

```python
# Data for the heatmap
np.random.seed(0)
data = np.random.rand(10, 10)

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Plot heatmaps with different colormaps
cmaps = ['viridis', 'plasma', 'inferno', 'magma']
for i, ax in enumerate(axs.flat):
    im = ax.imshow(data, cmap=cmaps[i])
    ax.set_title(cmaps[i])
    fig.colorbar(im, ax=ax)

# Layout so plots do not overlap
fig.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the violin plot
np.random.seed(0)
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 2
data3 = np.random.randn(100) + 4
data = [data1, data2, data3]

# Create a figure and axis
plt.figure(figsize=(8, 6))

# Plot the violin plot
plt.violinplot(data, showmeans=True, showextrema=True, showmedians=True)

# Set title and labels
plt.title('Violin Plot Example')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.xticks([1, 2, 3], ['Data 1', 'Data 2', 'Data 3'])

# Show the plot
plt.show()
```

**Task:** Add error bars to bar charts.

```python
# Data for the bar chart
np.random.seed(0)
x = np.arange(5)
y = np.random.rand(5)
y_err = np.random.rand(5) * 0.1

# Create a figure and axis
plt.figure(figsize=(8, 6))

# Plot the bar chart with error bars
plt.bar(x, y, yerr=y_err, capsize=5)

# Set title and labels
plt.title('Bar Chart with Error Bars')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(x, ['A', 'B', 'C', 'D', 'E'])

# Show the plot
plt.show()
```

### Exercise 46
**Task:** Customize plot borders/spines.

```python
# Data for the plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and axis
plt.figure(figsize=(8, 6))

# Plot the data
plt.plot(x, y)

# Customize the plot borders/spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('red')
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

# Set title and labels
plt.title('Customized Plot Borders')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

### Exercise 47
**Task:** Create a donut chart (pie chart with a hole).

```python
# Data for the donut chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create a figure and axis
plt.figure(figsize=(8, 8))

# Plot the donut chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', pctdistance=0.85, startangle=90)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax = plt.gca()
ax.axis('equal')  

# Set title
plt.title('Donut Chart Example')

# Show the plot
plt.show()

```
